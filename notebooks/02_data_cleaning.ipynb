# 🧹 Data Cleaning — Projeto de Preparação de Dados

Este notebook descreve as etapas de **limpeza e transformação** aplicadas ao dataset, garantindo que os dados estejam prontos para análise exploratória e modelagem preditiva.

---

### 1️⃣ Tratamento de Variáveis Binárias

As colunas com valores categóricos binários foram transformadas em **0 e 1** para facilitar o entendimento do modelo:

```python
# Transformação Y=1, N=0
df['CODE_GENDER'] = df['CODE_GENDER'].map({'Y':1, 'N':0})
df['FLAG_OWN_CAR'] = df['FLAG_OWN_CAR'].map({'Y':1, 'N':0})
df['FLAG_OWN_REALTY'] = df['FLAG_OWN_REALTY'].map({'Y':1, 'N':0})
```
---

⚡ Benefício: Facilita a interpretação binária pelo modelo e reduz erros de codificação manual.

---

### 2️⃣ Tratamento de Valores Faltantes (NaNs)

Para lidar com valores ausentes, aplicamos preenchimento específico por coluna:
```python
# Limpeza dos dados nans nas colunas [OCCUPATION_TYPE, years_employed]
df['OCCUPATION_TYPE'] = df['OCCUPATION_TYPE'].fillna('Missing')
df['years_employed'] = df['years_employed'].fillna(0)
```
---

⚡ Benefício: Evita perda de registros importantes e mantém a consistência do dataset.

---

### 3️⃣ Conversão de Colunas Categóricas

Colunas do tipo object foram convertidas para categorical, facilitando a interpretação do modelo sem a necessidade de one-hot encoding:

```python
# Transformando colunas categóricas em tipo category
cat_cols = [
    "NAME_INCOME_TYPE",
    "NAME_EDUCATION_TYPE",
    "NAME_FAMILY_STATUS",
    "NAME_HOUSING_TYPE",
    "OCCUPATION_TYPE",
]
for col in cat_cols:
    df[col] = df[col].astype('category')
```
---

⚡ Benefício: Compatível com categóricas nativas do XGBoost (enable_categorical=True).

---

### 4️⃣ Tratamento de Outliers

Foram definidos limites máximos para variáveis relacionadas à composição familiar, considerando que valores maiores seriam outliers:

```python
df['CNT_CHILDREN'] = df['CNT_CHILDREN'].clip(upper=6)
df['CNT_FAM_MEMBERS'] = df['CNT_FAM_MEMBERS'].clip(upper=7)
```
---

⚡ Benefício: Reduz o impacto de valores extremos, evitando distorções em métricas estatísticas e na modelagem.

---

### 5️⃣ Transformação Logarítmica

Variáveis de renda com distribuição assimétrica foram transformadas usando log1p, que permite lidar com valores zero:

```python
df['amt_income_month'] = np.log1p(df['amt_income_month'])
df['renda_per_capita'] = np.log1p(df['renda_per_capita'])
```

---

⚡ Benefício: Reduz a assimetria da distribuição, evita dominância de valores extremos e melhora a estabilidade dos splits em modelos baseados em árvore, como XGBoost..

---
### 6️⃣ Remoção de Colunas Irrelevantes

Algumas colunas não agregavam informação relevante para a modelagem:

```python
clients_fe = df.drop(columns=['ID', 'contact'], axis=1)
```

---

⚡ Benefício: Reduz dimensionalidade e melhora desempenho computacional.

---

## 7️⃣ Resumo do Data Cleaning
Etapa	Status
Transformação binária	✅ CODE_GENDER, FLAG_OWN_CAR, FLAG_OWN_REALTY
